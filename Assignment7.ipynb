```python
import numpy as np
import pandas as pd

roll_number = 102317054
np.random.seed(roll_number)
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
sales = np.random.randint(1000, 5001, (12, 4))
sales_df = pd.DataFrame(sales, columns=categories, index=months)
```

```python
print(sales_df.head())
print(sales_df.describe())
```

```python
total_sales_category = sales_df.sum()
print("Total sales per category:\n", total_sales_category)

sales_df_Total_Sales = sales_df.sum(axis=1)
print("Total sales per month:\n", sales_df_Total_Sales)
```

```python
average_growth = sales_df.pct_change().mean()
print("\nAverage sales growth between consecutive months for each category:")
print(average_growth)

sales_df_Growth_Rate = sales_df_Total_Sales.pct_change() * 100
```

```python
print("Percentage change in Total Sales from the previous month.:\n", sales_df_Growth_Rate)
```

```python
if roll_number % 2 == 0:
    sales_df['Electronics'] *= 0.9
else:
    sales_df['Clothing'] *= 0.85

print("Sales data after discount:")
print(sales_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sales_df[categories].plot(marker='o')
plt.title("Monthly Sales Trends")
plt.xlabel("Month")
plt.ylabel("Sales Units")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(7, 5))
sns.boxplot(data=sales_df[categories])
plt.title("Sales Distribution by Category")
plt.show()
```

```python
def swap_lst_elements(lst, idx1, idx2):
    temp = lst[idx1]
    lst[idx1] = lst[idx2]
    lst[idx2] = temp
    return lst

lst = [1, 2, 3, 4, 5]
print("Before Swap:", lst)
print("After Swap:", swap_lst_elements(lst , 1 , 4))
```
