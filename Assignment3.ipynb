```python
import pandas as pd
path="/content/drive/MyDrive/Dataset/employees.csv"
df=pd.read_csv(path)
avg_salary=df["Salary"].mean()
total_bonus=df["Bonus"].sum()
youngest_age=df["Age"].min()
highest_rating=df["Rating"].max()

print(f"Average Salary: {avg_salary}")
print(f"Average Salary: {total_bonus}")
print(f"Average Salary: {youngest_age}")
print(f"Average Salary: {highest_rating}")
```

```python
df_sorted=df.sort_values(by="Salary",ascending=False)
print(df_sorted)
```

```python
def categorize_performance(rating):
  if rating>=4.5:
    return "excellent"
  elif rating>=4:
    return "Good"
  else:
    return "average"

df["Performance_Category"]=df["Rating"].apply(categorize_performance)

print(df)
```

```python
missing_values = df.isnull().sum()

print(missing_values)
```

```python
df.rename(columns={"Employee_ID":"ID"})
```

```python
count=0
for index, row in df.iterrows():
    if row["Years_of_Experience"] > 5:
        count += 1
        print(f"Employee: {row['Name']}, Experience: {row['Years_of_Experience']} years")

print("\nTotal Employees with more than 5 years of experience:", count)
```

```python
count=0
for index, row in df.iterrows():
    if row["Department"] == 'IT':
        count += 1
        print(f"Employee: {row['Name']}, Department: {row['Department']}")

print("\nTotal Employees with Department as IT:", count)
```

```python
df["Tax"] = df["Salary"] * 0.10
print(df)
```
