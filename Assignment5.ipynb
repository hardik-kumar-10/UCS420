```python
X=ord('H')+ord('K')
sales=np.array([X+i*50 for i in range(5)])
tax_rate=((X%5)+5)/100
tax_sales=sales*tax_rate
discounted_sales = np.where(sales<X+100,sales*0.95, sales*0.90)
weekly_sales = np.tile(discounted_sales, (3, 1))
weekly_sales[1] *= 1.02
weekly_sales[2] *= 1.04
print('ASCII sum (X):',X)
print('Sales dataset:',sales)
print('Tax rate:',tax_rate)
print('Tax sales:',tax_sales)
print('Discounted_sales: ',discounted_sales)
print('Weekly sales: ',weekly_sales)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-10,10,100)
y1 = x**2
y2 = np.sin(x)
y3 = np.exp(x)
y4 = np.log(np.abs(x) + 1)

plt.figure(figsize=(10, 6))

plt.plot(x, y1, label="y = x^2", linestyle='--')
plt.plot(x, y2, label="y = sin(x)", linestyle='-.')
plt.plot(x, y3, label="y = e^x", linestyle=':')
plt.plot(x, y4, label="y = log(|x| + 1)", linestyle='-')
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Plots of Different Functions")
plt.legend()
plt.grid(True)
plt.show()
```
